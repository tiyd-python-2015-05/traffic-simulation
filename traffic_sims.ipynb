# Average Mean Speed and Standard Deviation after Nine Simulations

```python
speeds_of_nine = [nine_simulations[i][1][61:] for i in range(len(nine_simulations))]
mean_speeds_of_nine = [np.mean(speeds_of_nine[i]) for i in range(len(speeds_of_nine))]
stdev_of_nine = [np.std(speeds_of_nine[i]) for i in range(len(speeds_of_nine))]
```

```python
mean_of_mean_speeds_nine = np.mean(mean_speeds_of_nine)
```

```python
mean_of_stdev_nine = np.mean(stdev_of_nine)
```

```python
recommendation = mean_of_mean_speeds_nine + mean_of_stdev_nine
print(recommendation)
```

###Recommended Speed limit after Nine Simulations is 32 m/s

#Recommended Speed after 1000 Simulations

```python
simulations_1000 = n_simulations(1000)
speeds_1000 = [simulations_1000[i][1][61:] for i in range(len(simulations_1000))]
mean_speeds_1000 = [np.mean(speeds_1000[i]) for i in range(len(speeds_1000))]
stdev_1000 = [np.std(speeds_1000[i]) for i in range(len(speeds_1000))]
```

```python
"Slowest car speed is {}m/s and the fastest car spped is {}m/s".format(round(np.min(mean_speeds_1000), 1),round(np.max(mean_speeds_1000)), 1)
```

```python
plt.hist(mean_speeds_1000, bins=20)
plt.xlabel("Mean Speeds of Cars")
plt.ylabel("Frequency")
plt.title("Mean Speeds of 1000 Trials")
plt.subplots_adjust(left=9, right=10, top=10, bottom=9)
ymin, ymax = plt.ylim()
mean = np.mean(mean_speeds_1000)
stdev = np.mean(stdev_1000)
plt.vlines(mean, ymin, ymax, color="red")
plt.vlines([mean - stdev, mean + stdev], ymin, ymax, linestyles="dashed", color="red")
plt.show()
```

```python
"The average speed after 1000 trials is {} m/s".format(round(np.mean(mean_speeds_1000)))
```

##The recommended speed limit after 1000 simulations is 31.5m/s with an average car speed of 22 m/s
