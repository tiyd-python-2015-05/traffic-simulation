#Traffic Simulations of Driver Behavior on 1km  Roadway

I analyzed the behavior of drivers on a 1km length of road to determine the optimal speed limit. The number of cars on the roadway is first set to 30 and speed limit recommendations are compared with 1, 9 and 1000 simulations. Speed limit recommendations are then recalculated assuming the number of cars on the roadway varies between 20, 40, and 50. An additional 100 simulations are run with each variation of set number of cars.  added or removed to the 

```python
from simulation import Simulation
from car import Car
from simulation import n_simulations
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

##1. Traffic Simulations: 30 cars, 1 simulation

```python
#Setting car and simulation variables.
num_cars = 30
num_simulations = 1

#Calculating the mean speeds and standard deviation.
one_simulation = n_simulations(num_simulations, num_cars)
speed_of_one = [one_simulation[i][1][61:] for i in range(len(one_simulation))]
mean_speed_one = [np.mean(speed_of_one[i]) for i in range(len(speed_of_one))]
stdev_of_one = [np.std(speed_of_one[i]) for i in range(len(speed_of_one))]
print("{} cars, {} simulation:".format(num_cars, num_simulations))
print("  Mean speed: {} m/s".format(round(np.mean(mean_speed_one))))
print("  Standard Deviation: {}".format(round(np.mean(stdev_of_one))))
print("  Recommended speed limit is {} m/s"\
      .format(round(np.mean(mean_speed_one) + (round(np.mean(stdev_of_one))))))

##Scatter plot using plt.scatter
x = one_simulation[0][0]
y = [[i]*num_cars for i in range(121)]
plt.subplots_adjust(left=18, right=20, top=20, bottom=18)
plt.scatter(x, y, marker="|", c=['g', 'r', 'y', 'b'])
plt.xlabel("Length of Road (meters)", fontsize=18)
plt.ylabel("Time (seconds)", fontsize=18)
plt.title("Traffic Simulation: {} cars, {} simulation".format(num_cars, num_simulations), fontsize=15)
plt.hlines(60, 0, 1000, color='r')
plt.text(-175, 75, "Mean = {}".format(round(np.mean(mean_speed_one))), color='r', fontsize=14)
plt.text(-175, 70, "StDev = {}".format(round(np.mean(stdev_of_one))), color='r', fontsize=14)
plt.text(-175, 60, "Data Point", color='r', fontsize=14)
```

The redline indiactes the 60 second mark. The mean speed and standard deviation was calculated from 60s to 120s.
