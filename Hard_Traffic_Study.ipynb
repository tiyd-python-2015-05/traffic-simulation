#Do 100 Simulations to find average speed and standard deviation for speed limit recommendations.

```python
hundred_simulations = n_simulations(100)
```

```python
speeds = hundred_simulations[8][1][61:]
```

```python
speeds.shape
```

```python
speeds = [hundred_simulations[i][1][61:] for i in range(len(hundred_simulations))]
```

```python
mean_speeds = [np.mean(speeds[i]) for i in range(len(speeds))]
stdevs = [np.std(speeds[i]) for i in range(len(speeds))]
```

```python
print(np.mean(mean_speeds))
print(np.mean(stdevs))
```

#Based on the mean speed of 11.94 over 100 trials and the mean standard deviation of 11.82, the recommendation is to set the speed limit to 23.76 m/s.

The cars get into a huge traffic jam at km 3 when the braking chance is increased by 100%.  As they slowly get out of the jam, they have more room to accelerate and they can usually get up to a good cruising speed, because most of the cars are in a jam behind them.

This also happens at km 1 when braking is increased by 40%.  Sometimes the traffic jams are up to 1 km in length.  The average speed is not decreased as much as expected because of the acceleration that is possible after a huge traffic jam.
