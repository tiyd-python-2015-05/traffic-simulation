```python
from simulation_hard import Simulation
from car_hard import Car
from simulation_hard import n_simulations
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Setting car and simulation variables.
num_cars = 210
num_simulations = 1

#Calculating the mean speeds and standard deviation.
one_simulation = n_simulations(num_simulations, num_cars)
speed_of_one = [one_simulation[i][1][61:] for i in range(len(one_simulation))]
mean_speed_one = [np.mean(speed_of_one[i]) for i in range(len(speed_of_one))]
stdev_of_one = [np.std(speed_of_one[i]) for i in range(len(speed_of_one))]
print("{} cars, {} simulation:".format(num_cars, num_simulations))
print("  Mean speed: {} m/s".format(round(np.mean(mean_speed_one))))
print("  Standard Deviation: {}".format(round(np.mean(stdev_of_one))))
print("  Recommended speed limit is {} m/s"\
      .format(round(np.mean(mean_speed_one) + (round(np.mean(stdev_of_one))))))

##Scatter plot using plt.scatter
x = one_simulation[0][0]
y = [[i]*num_cars for i in range(481)]
plt.subplots_adjust(left=18, right=20, top=20, bottom=18)
plt.scatter(x, y, marker="|", c=['g', 'r', 'y', 'b'])
plt.xlabel("Length of Road (meters)", fontsize=18)
plt.ylabel("Time (seconds)", fontsize=18)
plt.title("Traffic Simulation: {} cars, {} simulation".format(num_cars, num_simulations), fontsize=15)
plt.vlines(1000, 0, 500, color='y')
plt.vlines(2000, 0, 500, color='y')
plt.vlines(3000, 0, 500, color='y')
plt.vlines(4000, 0, 500, color='y')
plt.vlines(5000, 0, 500, color='y')
plt.vlines(6000, 0, 500, color='y')
plt.text(1000, 590, "Braking chance 14%", rotation=45)
plt.text(2000, 590, "Braking chance 10%", rotation=45)
plt.text(3000, 590, "Braking chance 20%", rotation=45)
plt.text(4000, 550, "Braking normal", rotation=45)
plt.text(5000, 590, "Braking increased 12%", rotation=45)
plt.text(6000, 590, "Braking normal", rotation=45)
# plt.text(500, 500, "Mean = {}".format(round(np.mean(mean_speed_one))), fontsize=14)
# plt.text(-175, 70, "StDev = {}".format(round(np.mean(stdev_of_one))), color='r', fontsize=14)
# plt.text(-175, 60, "Data Point", color='r', fontsize=14)
```

```python
#Setting number of cars and number of simulations.
num_cars = 210
num_simulations = 100

#Mean and standard deviation only include time interval 61 - 481 seconds. 
simulations_100 = n_simulations(num_simulations, num_cars)
speeds_100 = [simulations_100[i][1][61:] for i in range(len(simulations_100))]
mean_speeds_100 = [np.mean(speeds_100[i]) for i in range(len(speeds_100))]
stdev_100 = [np.std(speeds_100[i]) for i in range(len(speeds_100))]

print("{} cars, {} simulations:".format(num_cars, num_simulations))
print("  Mean speed: {} m/s".format(round(np.mean(mean_speeds_100))))
print("  Standard Deviation: {}".format(round(np.mean(stdev_100))))
print("  Recommended speed limit is {} m/s".format(round(np.mean(mean_speeds_100) + np.mean(stdev_100))))
```

```python
#Range of car speeds 
print("Slowest car: {} m/s".format(round(np.min(mean_speeds_100))))
print("Fastest car: {} m/s".format(round(np.max(mean_speeds_100))))
```

```python
plt.hist(mean_speeds_100, bins=10)
plt.xlabel("Mean Speeds of Cars")
plt.ylabel("Frequency")
plt.title("Mean Speeds of 100 Trials")
plt.subplots_adjust(left=15, right=16, top=28, bottom=27)
ymin, ymax = plt.ylim()
mean = np.mean(mean_speeds_100)
stdev = np.mean(stdev_100)
plt.vlines(mean, ymin, ymax, color="red")
plt.vlines([mean - stdev, mean + stdev], ymin, ymax, linestyles="dashed", color="red")
# plt.text(0.0, 25, "Mean: {}".format(round(np.mean(mean_speeds_1000))), color='r') 
# plt.text(20, 120, "StDev: {}".format(round(np.mean(stdev_1000))), color='r')
plt.show()
```
